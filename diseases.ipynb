2. Explain the distribution of probability of any continuous variable.

```python
BMI_var = diseases['BMI']
BMI_vars = diseases['BMI'].unique()
(BMI_var)
```

```python
pb_BMI =[]



for i in BMI_vars:
   total =len(BMI_var[BMI_var == i])
   pb_BMI.append(total/len(diseases))

print(pb_BMI)

```

```python
BMI_event = np.array([BMI_vars, pb_BMI]).transpose()

BMI_events = pd.DataFrame(BMI_event, columns=['BMI_vars', 'pb_BMI'])
BMI_events
```

```python

sns.scatterplot(x = 'BMI_vars', y = 'pb_BMI', data=BMI_events)
```

2. Explain the distribution of probability of any continuous variable.
The data is normalized and positively skewed

5. Based on the variable “HeartDisease”, which "GenHealth" status  would you advise patients to maintain?

```python
heart_disease = diseases['HeartDisease']
Gen_health = diseases['GenHealth']

CrosstabResult = pd.crosstab(index=heart_disease,columns=Gen_health)
print(CrosstabResult)
CrosstabResult.plot.bar()
```

I would advise them to mantain an excellent GenHealth, because then, they will have few chances of acquiring heartDisease
